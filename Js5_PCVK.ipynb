### No. 1

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # Import cv2_imshow untuk menampilkan gambar di Google Colab

def gamma_correction(image, gamma):
    # Pastikan nilai gamma positif
    if gamma <= 0:
        raise ValueError("Nilai Gamma harus lebih besar dari 0")

    # Normalisasi nilai pixel ke dalam rentang [0, 1]
    image_normalized = image / 255.0

    # Lakukan Gamma Correction
    corrected_image = np.power(image_normalized, 1.0 / gamma)

    # Kembalikan ke rentang [0, 255]
    corrected_image = (corrected_image * 255).astype(np.uint8)
    return corrected_image

print(' Gamma Correction pada citra ')
print('----------------------------------')
try:
    gamma = float(input('Masukkan nilai Gamma: '))  # Mengubah menjadi float agar bisa menerima desimal
except ValueError:
    print('Error, bukan angka')

image = cv2.imread('female1.jpg')

# Pastikan citra berhasil dibaca
if image is None:
    print('Error: File citra tidak ditemukan atau tidak dapat dibaca.')
else:
    # Lakukan Gamma Correction
    gamma_corrected_image = gamma_correction(image, gamma)

    # Menggabungkan gambar asli dan hasil Gamma Correction secara horizontal
    final_frame = np.hstack((image, gamma_corrected_image))

    # Menampilkan gambar yang telah digabungkan menggunakan cv2_imshow
    cv2_imshow(final_frame)
```

### No. 2

```python
# Fungsi untuk melakukan kuantisasi citra dengan bit kedalaman tertentu
def quantize_image(image, bit_depth):
    # Hitung jumlah level (warna) yang mungkin berdasarkan bit kedalaman
    levels = 2 ** bit_depth

    # Hitung nilai level yang digunakan untuk kuantisasi
    level_value = 255 / (levels - 1)

    # Kuantisasi citra
    quantized_image = np.round(image / level_value) * level_value

    return quantized_image.astype(np.uint8)

# Bit kedalaman yang diinginkan
bit_depth = 2

# Baca citra asli
original = cv2.imread('female1.jpg', cv2.IMREAD_GRAYSCALE)

# Lakukan kuantisasi citra
depth_image = quantize_image(original, bit_depth)

# Tampilkan citra asli dan citra kuantisasi
final_frame = np.hstack((original, depth_image))

# Menampilkan gambar yang telah digabungkan menggunakan cv2_imshow
cv2_imshow(final_frame)
```
